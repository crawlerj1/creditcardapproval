```python

def remove_outliers(df, columns):
    for column in columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        # Define the lower and upper bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filter out the outliers
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df
```

```python
traindata=remove_outliers(traindata,['Income', 'Employment length','Family member count','Children count'])
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, MinMaxScaler,LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
```

```python
employmentsta_categories = ['Student','Working','Commercial associate', 'State servant','Pensioner']
education_categories = ['Lower secondary','Secondary ','Incomplete higher', 'Higher education', 'Academic degree']
marital_categories = ['Single ','Married','Civil marriage','Separated','Widow']
dwelling_categories = ['With parents','House ','Co-op apartment','Municipal apartment','Rented apartment','Office apartment']
job_categories=['Core staff',
 'Accountants',
 'Laborers',
 'Managers',
 'Sales staff',
 'Medicine staff',
 'High skill tech staff',
 'HR staff','Low-skill Laborers','Drivers',
 'Secretaries','Cleaning staff','Cooking staff',
 'Security staff','Private service staff','IT staff','Waiters','Realty agents']
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf2',OrdinalEncoder(categories=[['Secondary ', 'Higher education', 'Lower secondary',
       'Incomplete higher', 'Academic degree']]),['Education level']),
    ('tnf3',OneHotEncoder(sparse_output=False),['Gender', 'Has a car', 'Has a property', 'Children count',
       'Employment status',  'Marital status', 'Dwelling', 'Job title'])
],remainder='passthrough')
```

```python
X_train=transformer.fit_transform(X_train)
X_test=transformer.transform(X_test)
```

```python
transformer2=ColumnTransformer(transformers=[
    ('tnf3',MinMaxScaler(),slice(0,36))
])
```

```python
X_train=transformer2.fit_transform(X_train)
X_test=transformer2.transform(X_test)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_balance, y_balance = smote.fit_resample(X_train,y_train)
X_test, y_test = smote.fit_resample(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,precision_score,recall_score
```

```python
log_model = LogisticRegression()
log_model.fit(X_balance, y_balance)
y_pred = log_model.predict(X_test)
print(f'Accuracy score LR: {accuracy_score(y_test,y_pred)}')
print(f'Precision LR: {precision_score(y_test,y_pred)}')
print(f'recall score LR: {recall_score(y_test,y_pred)}')
print(f'confusion matrix LR:\n {confusion_matrix(y_test,y_pred)}')
```

```python
desmodel =DecisionTreeClassifier()
desmodel.fit(X_balance, y_balance)
y_pred2= desmodel.predict(X_test)
print(f'Accuracy score DT: {accuracy_score(y_test,y_pred2)}')
print(f'Precision DT: {precision_score(y_test,y_pred2)}')
print(f'recall score DT: {recall_score(y_test,y_pred2)}')
print(f'confusion matrix DT:\n {confusion_matrix(y_test,y_pred2)}')
```

```python
randmodel =RandomForestClassifier(random_state=42,n_estimators=100)
randmodel.fit(X_balance, y_balance)
y_pred3 = randmodel.predict(X_test)
print(f'accuracyscore RF: {accuracy_score(y_test,y_pred3)}')
print(f'Precision RF: {precision_score(y_test,y_pred3)}')
print(f'recall score RF: {recall_score(y_test,y_pred3)}')
print(f'confusion matrix RF:\n {confusion_matrix(y_test,y_pred3)}')
```

```python
plt.figure(figsize=(2,2))
sns.set_style('white')
sns.heatmap(confusion_matrix(y_test,y_pred3), annot=True, fmt='d', cmap='PuBuGn_r', cbar=False)
plt.title('Confusion Matrix: Random Forest')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
